## Negative of an Image
In this part, you will have to implement and display the negative of the image. in both Gray, and RGB spaces.
do not use built-in functions.

````python
#TODO
# Negative of RGB image
def neg_RGB(img):
    w, h, _ = img.shape
    for i in range(w):
        for j in range(h):
            img[i][j][0] = 255 - img[i][j][0]
            img[i][j][1] = 255 - img[i][j][1]
            img[i][j][2] = 255 - img[i][j][2]
    
    return img
tmp = RGB_image.copy()
cv2.imshow("RGB-negative", neg_RGB(tmp))
cv2.waitKey(0)
cv2.destroyAllWindows()

#Negative of Gray image
def neg_gray(img):
    w, h = img.shape
    for i in range(w):
        for j in range(h):
            img[i][j] = 255 - img[i][j]
    
    return img

tmp = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("gray-negative", neg_gray(tmp))
cv2.waitKey(0)
cv2.destroyAllWindows()
````

## Binarize Image
 Convert the input image to Gray, and then binarize it. do not use built-in functions.

````python
#TODO
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
w, h = gray_image.shape
print(gray_image.shape)
binarized_image = gray_image.copy()
threshold = 255 / 2
for i in range(w):
    for j in range(h):
        if gray_image[i][j] > threshold:
            binarized_image[i][j] = 255
        else:
            binarized_image[i][j] = 0

result = np.concatenate((gray_image, binarized_image), axis = 1)

cv2.imshow("gray cat vs binarized cat", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

## Bit-Plane Slicing


Perform Bit-Plane Slicing on the input image.
Implement the function yourself, do not use built-in functions.

````python
#TODO  BitPlane Slicing
def bitplane_slicing(img):
    res = []
    w, h = img.shape
    for i in range(0, 8):
        bit = np.full((w, h), 2 ** i, dtype=np.uint8)
        tmp = np.bitwise_and(img, bit)
        tmp = np.multiply(tmp, 255)
        res.append(tmp)
    res = np.array(res, dtype=np.uint8)
    return res

result = bitplane_slicing(gray_image)
tmp1 = np.concatenate((result[0], result[1], result[2], result[3]), axis=1)
tmp2 = np.concatenate((result[4], result[5], result[6], result[7]), axis=1)
tmp3 = np.concatenate((tmp1, tmp2), axis=0)
cv2.imshow("bit plane sliced", tmp3)
cv2.waitKey(0)
cv2.destroyAllWindows()

'''if all of the images don't fit in your screen you can run the code below to see all of the bit-planes separately'''
# for i in range(0, 8):
#     cv2.imshow(f"{i}-bit plane sliced", result[i])
#     cv2.waitKey(0)
#     cv2.destroyAllWindows()
````

As you can see, cv2.cvtColor() function converts colorspace.
You have to write your own function RGB_to_HSV without using opencv and other built-in functions, to convert an RGB image to an HSV one. compare your results with the above code block.

````python
#TODO  def RGB_to_HSV()
def RGB_to_HSV(RGB_img):
    w, h, _ = RGB_img.shape
    red = RGB_img[:, :, 0]
    green = RGB_img[:, :, 1]
    blue = RGB_img[:, :, 2]
    res = [[0 for _ in range(h)] for _ in range(w)]
    for i in range(w):
        for j in range(h):
           res[i][j] = round(0.3*red[i][j] + 0.59*green[i][j] + 0.11*blue[i][j])
    res = np.array(res, dtype=np.uint8)
    return res

result = RGB_to_HSV(RGB_image)

cv2.imshow("rgb to hsv", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

## Histogram Equalization
In this section you will have to perform histogram equalization on the gray image. This transform flattens the gray-level histogram so that all intensities are as equally common as possible. The transform function is a cumulative distribution function (cdf) of the pixel values in the image (normalized to map the range of pixel values to the desired range).

To do so,
1. first display the original histogram of the image using built-in functions,
2. Then write a function to perform Histogram Equalization without using any built-in function, only Numpy.
3. Use the cv2.equalizeHist to compare your results.

````python
# TODO : def hist_eq()
hist = np.array(cv2.calcHist([gray_image],[0],None,[256],[0,256]))
total_pixels = np.sum(hist)
cdf = np.full((len(hist)), 0, dtype=np.float32)
sum = 0
mapping = []
for i in range(len(hist)):
    sum += hist[i]
    cdf[i] = (sum / total_pixels)
    mapping.append(round((len(hist)-1) * cdf[i]))
    
def hist_eq(gray_img):
    w, h = gray_img.shape
    equalized_img = gray_image.copy()
    for i in range(w):
        for j in range(h):
            equalized_img[i][j] = mapping[gray_image[i][j]]
    
    return np.array(equalized_img, dtype=np.uint8)

manual_equalized_image = hist_eq(gray_image)
    
cv2_equalized_image = cv2.equalizeHist(gray_image)
result = np.hstack((manual_equalized_image, cv2_equalized_image))

cv2.imshow("manual hist equalization vs cv2 hist equalization", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
````
